```python
from pylab import *
from numpy.matlib import *

def report_layout(size=(12,9), latex = False, svg = True):
    """
    Changes global matplotlib settings to make plots nicer. Size in centimeters.
    """
    from IPython.display import set_matplotlib_formats
    %matplotlib inline
    
    set_matplotlib_formats('svg' if svg else 'png', 'pdf')
    params = {'legend.fontsize': 12,
              'figure.figsize': (size[0] / 2.54, size[1] / 2.54),
              'axes.labelsize': 12,
              'axes.titlesize': 12,
              'xtick.labelsize': 10.95,
              'ytick.labelsize': 10.95,
              'axes.spines.left': False,
              'axes.spines.right': False,
              'axes.spines.top': False,
              'axes.spines.bottom': False,
              'axes.axisbelow': True,
              'axes.grid': True,
              'grid.color': '0.5',
              'grid.alpha': '0.5',
              'text.usetex': latex,
              'font.family': 'serif',
              'font.serif': ['Computer Modern', 'Bitstream Vera Serif'],
              'xtick.major.width': 0,
              'ytick.major.width': 0,
              'legend.frameon': False,
              'savefig.bbox': 'tight'
             }
    rcParams.update(params)
```
